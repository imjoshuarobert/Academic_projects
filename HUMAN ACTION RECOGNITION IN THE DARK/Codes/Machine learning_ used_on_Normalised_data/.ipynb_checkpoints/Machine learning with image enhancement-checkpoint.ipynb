# Image enhancement

```python

import cv2
import numpy as np

def gamma_correction(image, gamma=1.0):
    # Build a lookup table mapping the pixel values [0, 255] to their adjusted gamma values
    inv_gamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** inv_gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    return cv2.LUT(image, table)

def adjust_brightness_contrast(frame, alpha=1.0, beta=0):
    # Apply brightness and contrast adjustment
    adjusted_frame = cv2.convertScaleAbs(frame, alpha=alpha, beta=beta)
    return adjusted_frame

def adaptive_gamma_correction_and_brightness_contrast(video_path, output_path, initial_gamma=1.5, min_gamma=0.5, max_gamma=2.0, step=0.1, brightness=0, contrast=1.0):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video file.")
        return
    
    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Process each frame
    frame_number = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Apply adaptive gamma correction to the frame
        gamma_value = initial_gamma + (frame_number / total_frames) * (max_gamma - min_gamma)
        gamma_corrected_frame = gamma_correction(frame, gamma_value)

        # Apply adaptive brightness and contrast adjustment
        adjusted_frame = adjust_brightness_contrast(gamma_corrected_frame, alpha=contrast, beta=brightness)

        # Write the adjusted frame to output video
        out.write(adjusted_frame)

        # Display progress
        print("Processed frame {}/{}".format(frame_number + 1, total_frames), end='\r')

        frame_number += 1

    # Release resources
    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print("\nAdaptive gamma correction and brightness/contrast adjustment completed.")

```

```python
import cv2
import numpy as np
from keras.applications import InceptionV3
from keras.applications.inception_v3 import preprocess_input

# Function to extract features from frames using InceptionV3
def extract_features(frames):
    # Load InceptionV3 model (excluding top layers)
    base_model = InceptionV3(weights='imagenet', include_top=False, pooling='avg')
    
    # Preprocess frames
    preprocessed_frames = preprocess_input(np.array(frames))
    
    # Extract features using InceptionV3
    features = base_model.predict(preprocessed_frames)
    
    return features


def process_video(video_path, max_frames=25):
    # Read video
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Calculate sampling interval
    sampling_interval = max(1, frame_count // max_frames)
    
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Resize frame to 299x299
        frame = cv2.resize(frame, (299, 299))
        
        frames.append(frame)
        # Skip frames based on sampling interval
        for _ in range(sampling_interval - 1):
            cap.grab()
    
        # Break if maximum frames reached
        if len(frames) == max_frames:
            break
    
    cap.release()
    
    # Pad frames if necessary
    if len(frames) < max_frames:
        frames.extend([frames[-1]] * (max_frames - len(frames)))
    
    # Extract features
    features = extract_features(frames)
    
    return features
```

# Metric 

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import seaborn as sns
#generates a confusion matrix between hand labelled data and model predictions
def getConfMatrix(pred_data, actual):
    conf_mat = confusion_matrix(actual, pred_data, labels=[0,1]) 
    micro = f1_score(actual, pred_data, average='micro') 
    macro = f1_score(actual,pred_data, average='macro')
    sns.heatmap(conf_mat, annot = True, fmt=".0f", annot_kws={"size": 18})
    print('F1 Micro: '+ str(micro))
    print('F1 Macro: '+ str(macro))
```

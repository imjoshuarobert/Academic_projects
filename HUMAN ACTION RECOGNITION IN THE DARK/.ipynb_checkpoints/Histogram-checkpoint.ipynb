# Histogram 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show_frame_and_histogram(frame):
    """
    Display both the frame and its histogram side by side.
    
    Args:
    - frame: numpy array representing the video frame
    """
    # Convert frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Calculate histogram
    histogram = cv2.calcHist([gray_frame], [0], None, [256], [0, 256])
    
    # Create subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    
    # Display frame
    ax1.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    ax1.set_title('Sample Frame')
    ax1.axis('off')
    
    # Plot histogram
    ax2.plot(histogram, color='black')
    ax2.set_xlabel('Pixel Intensity')
    ax2.set_ylabel('Frequency')
    ax2.set_title('Histogram of First Frame')
    
    # Show plots
    plt.show()

def extract_first_frame(video_path):
    """
    Extract the first frame from a video file.
    
    Args:
    - video_path: path to the input video file
    
    Returns:
    - first_frame: numpy array representing the first frame
    """
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
    
    # Read the first frame
    ret, first_frame = video_capture.read()
    
    # Release video capture object
    video_capture.release()
    
    return first_frame

# Example usage:
input_video_path = '1.mp4'  # Replace with your input video file path

# Extract the first frame
first_frame = extract_first_frame(input_video_path)

# Display the sample frame and its histogram
show_frame_and_histogram(first_frame)
```
